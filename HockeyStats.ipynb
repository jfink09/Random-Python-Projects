```python
import pandas as pd
```

```python
data1 = pd.ExcelFile("nhl11-12.xlsx")
data2 = pd.ExcelFile("nhl11-13.xlsx")

df1 = pd.read_excel(data1)
df2 = pd.read_excel(data2)
```

```python
df1
```

```python
df2
```

```python
GF1 = df1["GF"]
GF2 = df2["GF"]

print(f"GF From Second To Last Game: {GF1}")
print(f"GF From Last Game: {GF2}")
```

```python
GA1 = df1["GA"]
GA2 = df2["GA"]

print(f"GA From Second To Last Game: {GA1}")
print(f"GA From Last Game: {GA2}")
```

```python
# Total = GF + GA
total_1 = GF1 + GA1
total_2 = GF2 + GA2

print(f"Total From Second To Last Game: {total_1}")
print(f"Total From Last Game: {total_2}")
```

```python
# Spread = GF - GA
spread_1 = GF1 - GA1
spread_2 = GF2 - GA2

print(f"Spread From Second To Last Game: {spread_1}")
print(f"Spread From Last Game: {spread_2}")
```

```python
# Spread = GF - GA
points_1 = df1["P"]
points_2 = df2["P"]
teams_1 = df1["Team"]
teams_2 = df2["Team"]

print(f"Points From Second To Last Game: {points_1}")
print(f"Points From Last Game: {points_2}")
```

```python
df1["Table"] = teams_1.astype(str) +"             "+ points_1.astype(str)
df1["Table"]
```

```python
df2["Table"] = teams_2.astype(str) +"             "+ points_2.astype(str)
df2["Table"]
```
